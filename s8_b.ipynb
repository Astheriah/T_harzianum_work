```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('resultados_1.csv', sep=',')  

df.head()
```

```python

sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)



# ---------- Gráfico 1: Genes faltantes (menor a mayor) ----------
plt.figure(figsize=(10, 6))
df_sorted = df.sort_values('PorcentajeFaltantes', ascending=True)
ax = sns.barplot(data=df_sorted, x='Modelo', y='PorcentajeFaltantes', palette='Reds')
plt.title('Porcentaje de genes faltantes', fontsize=14)
plt.ylabel('% genes faltantes')
plt.xlabel('Modelo')
plt.xticks(rotation=45)

# Etiquetas de datos
for p in ax.patches:
    h = p.get_height()
    if h > 0:
        ax.text(p.get_x() + p.get_width()/2., h + 0.5, f'{h:.1f}%', 
                ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()

# ---------- Gráfico 2: Reacciones no presentes (menor a mayor, al revés) ----------
plt.figure(figsize=(10, 6))
df_sorted = df.sort_values('PorcentajeNoPresentes', ascending=True)  # <--- Ascendente
ax = sns.barplot(data=df_sorted, x='Modelo', y='PorcentajeNoPresentes', palette='Reds')
plt.title('Porcentaje de rxns no presentes', fontsize=14)
plt.ylabel('% rxns no presentes')
plt.xlabel('Modelo')
plt.xticks(rotation=45)

for p in ax.patches:
    h = p.get_height()
    if h > 0:
        ax.text(p.get_x() + p.get_width()/2., h + 0.5, f'{h:.1f}%', 
                ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()

# ---------- Gráfico 3: Reacciones completas (mayor a menor) ----------
plt.figure(figsize=(10, 6))
df_sorted = df.sort_values('PorcentajeCompletas', ascending=False)
ax = sns.barplot(data=df_sorted, x='Modelo', y='PorcentajeCompletas', palette='Greens_r')
plt.title('Porcentaje de rxns', fontsize=14)
plt.ylabel('% rxns')
plt.xlabel('Modelo')
plt.xticks(rotation=45)

for p in ax.patches:
    h = p.get_height()
    if h > 0:
        ax.text(p.get_x() + p.get_width()/2., h + 0.5, f'{h:.1f}%', 
                ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()
```
